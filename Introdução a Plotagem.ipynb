Iremos inicialmente usar o quintessencial para visualização dos dados, o quarteto de Anscombe

Os dados de Anscombe comtêm quatro conjuntos de dados, cada um com duas variáveis contínuas. Todos os conjuntos
têm a mesma média, variância, Correlação e linha de regressão. No entanto, somente quando os dados são visualizados
é q se torna óbvio que os conjuntos não seguem o mesmo padrão. Isso serve para mostrar as vantagens da visualizações 
e as armadilhas de se observarem somente estatisticas resumidas.



```python
# O conjunto de dados de Anscombe pode ser encontrado na biblioteca seaborn
import seaborn as sns
anscombe = sns.load_dataset("anscombe")
print(anscombe.head())
```

```python
#MatpLotlib! para uso dessa biblioteca, sua importação é diferente, onde todos os utilitários podem ser encontrador em uma 
# "subpasta" chamada pyplot.
#obs para uso dessa biblioteca no jupter notebook é necessário inserir "% matplotlib notebook"
import matplotlib.pyplot as plt
```

```python
# Os nomes da maioria das plotagens básicas começará com plt.plot
# criar um subconjunto de dados
# contém somente o primeiro conjunto de dados de Anscombe
dataset_1 = anscombe[anscombe['dataset'] == 'I']

%matplotlib notebook
plt.plot(dataset_1['x'], dataset_1['y'])
```

Por padrão plt.plot desenhará linhas, mas se quiser plotar círculos (pontos), para isso basta passar o parâmetro 'o'
para dizer ao plt.plot que use a figura na plitagem.

```python
%matplotlib notebook 
# não se esqueça disso
plt.plot(dataset_1['x'], dataset_1['y'], 'o')
```

A sintaxe de subplot aceita três parâmetros:

1) Número de linhas na figura para as subplotagens <br>
2) Número de colunas na figura para as subplotagens <br>
3) Local da subplotagem



```python
#Criando subconjunto dos dados de Anscombe

dataset_2 = anscombe[anscombe['dataset']=='II']
dataset_3 = anscombe[anscombe['dataset']=='III']
dataset_4 = anscombe[anscombe['dataset']=='IV']


%matplotlib notebook
# Criando figura na qual nossas subplotagens serão inseridas
fig = plt.figure() # Diz a figura como as subplotagens deverão ser dispostas no exemplo,
                   # Teremos 2 linhas de plotagens 2 duas colunas, local 1 da plotagem                 
axes1 = fig.add_subplot(2, 2, 1)

# A subplotagem tem 2 duas linhas e 2 colunas, local 2 da plotagem
axes2 = fig.add_subplot(2, 2, 2)

# A subplotagem tem 2duas linhas e 2 colunas, local 3 da plotagem
axes3 = fig.add_subplot(2, 2, 3)

# A subplotagem tem duas linhas e duas colunas, local 4 da plotagem
axes4 = fig.add_subplot(2, 2, 4)

#Podemos usar o método plot em cada eixo para criar nossa plotagem
#Adiciona uma plotagem em cada eixo criado acima

axes1.plot(dataset_1['x'], dataset_1['y'], 'o')
axes2.plot(dataset_2['x'], dataset_2['y'], 'o')
axes3.plot(dataset_3['x'], dataset_3['y'], 'o')
axes4.plot(dataset_4['x'], dataset_4['y'], 'o')

# Incluindo titulo ara cada subplotagem
axes1.set_title('dataset_1')
axes2.set_title('dataset_2')
axes3.set_title('dataset_3')
axes4.set_title('dataset_4')

# Adiciona um titulo para toda a figura
fig.suptitle("Anscombe Data")

#Organizando layout
fig.tight_layout()
```

# Univariado!


No jargão de estatistica univariado refere-se a uma única variável.


```python
'''
Os histogramas são o meio mais comum de oservar uma única variável. Os valores são colocados em 'recipientes', isto é,
são agrupados e plotados para ostrar a distribuição da variável
'''
import matplotlib.pyplot as plt
%matplotlib notebook

fig = plt.figure()
axes1 = fig.add_subplot(1, 1, 1)
axes1.hist(tips['total_bill'], bins=10)
axes1.set_title('Histograma do total de conta')
axes1.set_xlabel('Frequencia')
axes1.set_ylabel('Conta total')
fig.show()
```

# Bivariado

No jargão do estatística o termo 'bivariado' refere-se a duas variáveis

```python
# Gráficos de dispersão
# Gráficos de dispersão (scatterplot) são usados quando uma variável contínua 
# é plotada em relação a outra variável contínua.

%matplotlib notebook
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1, 1, 1)
axes1.scatter(tips['total_bill'], tips['tip'])
axes1.set_title('Dispersão do total de conta vs gorjeta')
axes1.set_xlabel('Conta Total')
axes1.set_ylabel('Gorjeta')
scatter_plot.show()
```

```python
# Gráfico de Caixa
# Boxplot são usados quando uma variavel discreta é plotada em relação a outra variável contínua.
%matplotlib notebook
boxplot = plt.figure()
axes1 = boxplot.add_subplot(1, 1, 1)
axes1.boxplot(
    # o primeiro argumento de um gráfico de caixa é dado,
    # pois estamos plotando várias porções de dados,
    # temos que colocar cada porção de dados em uma lista
    [tips[tips['sex'] == 'Female']['tip'],
     tips[tips['sex'] == 'Male']['tip']],
    
    # podeos então passar um parâmetro labels opcional
    # para especificar um rótulo aos dados que passamos
    labels=['Mulher', 'Homem'])

axes1.set_xlabel('Sexo')
axes1.set_ylabel('Gorjeta')
axes1.set_title('Boxplot de gorjeta por sexo')
plt.show()
```

# Dados multivariados

Para reduzir a confusão dada na plotagem de dados multivariados, devido a sobreposição dos dados, pode ser sanada
introduzindo uma certa dose de transparência aos pontos individuais, de modo que muitos pontos sobrepostos serão exibidos
como uma região mais escura de uma plotagem, em comparação com áreas menos cheias.


```python
#Cria uma variável de cor baseada no sexo
%matplotlib notebook

def recode_sex(sex):
    if sex == 'Female':
        return 0
    else:
        return 1
    
tips['sex_color'] = tips['sex'].apply(recode_sex)

scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1, 1, 1)
axes1.scatter(
    x=tips['total_bill'],
    y=tips['tip'],
    
    # Define o tamanho dos pontos com base no tamanho dos grupos;
    # multiplicamos os valore spo 10 para deixarem os pontos maiores
    # e enfatizar a diferença
    s=tips['size']*10,
    
    # Define valor apra sexo
    c=tips['sex_color'],
    
    # Define o valor de alpha para que os pontos seham mais tranparentes;
    # isso ajuda no caso de pontos que se sobrepoem
    alpha=0.5)

axes1.set_title('Contas x Gorjeta x sexo')
axes1.set_xlabel('Total de contas')
axes1.set_ylabel('Gorjeta')
scatter_plot.show()
```

# Plotagem de contadores
As plotagens de barra (bar plot) são muito parecidas com os histogramas, mas, <br> 
em vez de colocar os valores "em recepientes" para gerar uma distribuição, as <br>
plotagens de barra podem ser usdas para contar variáveis discretas. uma plotagem <br>
de contadores (cont plot) é usada com essa finalidade

```python
%matplotlib notebook
count, ax = plt.subplots()
ax = sns.countplot('day',data=tips)
ax.set_title('Contador de dias')
ax.set_xlabel('Dias da semana')
ax.set_ylabel('Frequencia')
plt.show()
```

```python
# Dados bivariados
# Cria uma gráfico de dispersão, além de criar uma linha de regressão.
# Se definirmos 'fit_reg=False' ela retornará somente o gráfico de dispersão.

%matplotlib notebook
scatter, ax = plt.subplots()
ax = sns.regplot(x='total_bill', y='tip', data=tips)
ax.set_title('Scaterplot do total de contas e Gorjetas')
ax.set_xlabel('total de contas')
ax.set_ylabel('Gorjeta')
plt.show()
```

Os gráficos de dispersão são ótimos para comparar duas variáveis. <br> 
Ás vezes, porém, há pontos demais para que um gráfico de dispersão seja significativo <br>
Uma forma de constornar esse problema e reunir pontos no gráfico. Assim como os histogramas <br>
podem reunir dados de uma variável para criar uma barra, o 'hexbin' pode fazer o mesmo<br>
com duas variáveis. Hum exhágono é usado com essa finalidade, pois é o formato mais eficiente<br>
para cobrir uma superfície 2D arbitrária. Esse é um exemplo do seaborn funcionando com base<br>
no matplotlib, pois o hexbin é uma função do matplotlib.


```python
%matplotlib notebook

hexbin = sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
hexbin.set_axis_labels(xlabel='Conta total', ylabel='Gorjeta')
hexbin.fig.suptitle('Hexbin Joint Plot of Total bill ant Tip', fontsize=10, y=1.03)
```

Também é possível criar uma plotagem de densidade por kernel 2D.
Esse tipo de processo é semante ao modo como sns.kedplot funciona, exeto por criar uma densidade 
entre duas variáveis. A plotagem bivariada pode ser mostrada sozinha, ou você pode colocar as duas 
plotagens univariadas uma ao lado da outra usando jointplot.

```python
kde, ax = plt.subplots()
ax = sns.kdeplot(data=tips['total_bill'],
                data2=tips['tip'],
                shade=True) #Sobreamento preencherá os contornos
ax.set_title('kernel Density Plot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
plt.show()
```

```python
kde_joint = sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
```

# Plotagem de barras
As plotagens de barras (bar plots) também podem ser usadas para mostrar diversas variáveis. Por padrão, barplot calculará
uma média, mas voce pode passar qualquer função para o parâmetro estimador. Por exemplo, poderiamos passar a a função 
numpy.std para calcular o desvio padrão.

```python
bar, ax = plt.subplots()
ax = sns.barplot(x='time', y='total_bill', data=tips)
ax.set_title('Bar plot of avarege total bill for time of days')
ax.set_xlabel('Time of day')
ax.set_ylabel('Average total bill')
plt.show()
```

```python
# Gráfico de caixa
'''
Um platagem de caixa (boxplot) mostra várias estatísticas: O mínimo, o primeiro quartil, a  mediana, 
o terceiro quartil, o maximo e, se for aplicável, valores discrepantes baseados no intervalo interquartil.
'''
# O parâmetro y é opcional, Se for omitido, a função de plotagem criará uma única caia no gráfico

box, ax = plt.subplots()
ax = sns.boxplot(x='time', y='total_bill', data=tips)
ax.set_title('Boxplot do total de contas por dias da semana')
ax.set_xlabel('Dias da semana')
ax.set_ylabel('Conta total')
plt.show()
```

```python
#Violin
violin, ax = plt.subplots()
ax = sns.violinplot(x='time', y='total_bill', data=tips)
ax.set_title('Violin plot do total de venda por dias da semana')
ax.set_xlabel('Dia da semana')
ax.set_ylabel('Total vendas')
plt.show()
```

```python
# Relacionamento aos pares
'''
Quando temos dados numéricos em sua maioria, a visualização de todos os relacionaento aos pares pode ser feita facilmente
usando pairplot. Essa função plotará um gráfico de dispersão entre cada par de variáveis, e um histograma para os dados
univariados.
'''
fig = sns.pairplot(tips)
```

```python
'''
Uma desvantagem ao usar pairplot está no fato de haver informações redundantes, istp é, a parte superior da visualização
é igual a metade inferior. Podemos usar pairgrid para atribuir manualmente as plotagens para a metade superior e
a metade inferior.
'''
# podemos usar plt.scatter em vez de sns.regplot
pair_grid = sns.PairGrid(tips)
pair_grid = pair_grid.map_upper(sns.regplot)
pair_grid = pair_grid.map_lower(sns.kdeplot)
pair_grid = pair_grid.map_diag(sns.distplot, rug=True)
plt.show()
```

```python
# Cores
# Passando o parâmetro hue para colorir a plotagem de acordo com o sex.
violin, ax = plt.subplots()
ax = sns.violinplot(x='time', y='total_bill',
                   hue='sex', data=tips,
                   split=True)
plt.show()
```

```python
# O parametro hue pode ser passado para várias outras funções de plotagem tabm.
# Observe o uso de Lmplot no lugar de regplot aqui
scatter = sns.lmplot(x='total_bill', y='tip', data=tips,
                    hue='sex', fit_reg=False)
plt.show()
```

```python
'''
Podemos deixar nossas plotagens de pares um pouco mais significativas passando uma das variaveis de categoria como
o parametro hue.
'''
fig = sns.pairplot(tips, hue='sex')
```

# Tamanho e formato

Trabalhar com tamano de pontos pode ser uma outra forma de acrescer mais informações em uma plotagem. No entanto, essa opção deve ser usada com moderação, pois o olho humano não é muito bom para comparar áreas.

Considere um exemplo de como o seaborn trabalha com chamadas de função do matplotlib. Se consultar a codumentação dom lmplot, voce verá que ela acira parametros de nomes catter, line scatter, line_kws. Os dois parametros acitam um par chave-valor- um dicionario puthonpara ser mais exato.


```python
scatter = sns.lmplot(x='total_bill', y='tip', data=tips,
                    fit_reg=False,
                    hue='sex',
                    scatter_kws={'s': tips['size']*10})
plt.show()
```

```python
scatter = sns.lmplot(x='total_bill', y='tip', data=tips,
                    fit_reg=False,
                    hue='sex',
                    markers=['o','x'],
                    scatter_kws={'s': tips['size']*10})
plt.show()
```
